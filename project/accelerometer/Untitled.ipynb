```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('value.csv',header=None)
```

```python
df
```

```python
df_standing = df[df[3] == 0]
```

```python
df_standing.head()
```

```python
df_standing.describe()
```

```python
plt.plot(df_standing[0],label='x')
plt.plot(df_standing[1],label='y')
plt.plot(df_standing[2],label='z')
plt.legend()
```

```python
df_sitting = df[df[3] == 1]
```

```python
plt.plot(df_sitting[0],label='x')
plt.plot(df_sitting[1],label='y')
plt.plot(df_sitting[2],label='z')
plt.legend()
```

```python
plt.plot(df_standing[2],label = 'standing')
plt.plot(df_sitting[2],label='sitting')
plt.legend()
```

```python
plt.plot(df_standing[1],label = 'standing')
plt.plot(df_sitting[1],label='sitting')
plt.legend()

```

```python
plt.plot(df_standing[0],label = 'standing')
plt.plot(df_sitting[0],label='sitting')
plt.legend()

```
